## Q1: Define language processing and list all compiler phases with explanations.

### DETAILED ANSWER:

**Definition of Language Processing:**

Language processing is the systematic conversion of a program written in a source language into an equivalent program in a target language, while maintaining the semantic meaning and functionality of the original program. This process bridges the gap between human-readable code and machine-executable instructions.

**The Seven Main Compiler Phases:**

#### 1. **Lexical Analysis (Scanning)**
- **What it does:** Reads source code character-by-character and groups them into meaningful units called **tokens**
- **Tokens created:** Keywords (if, while, int), Identifiers (variable names), Numbers (constants), Operators (+, -, *), Special symbols (;, {, })
- **Example:**
  - Input: `int count = 5 + 3;`
  - Tokens: KEYWORD(int), IDENTIFIER(count), ASSIGN(=), NUMBER(5), PLUS(+), NUMBER(3), SEMICOLON(;)
- **Error handling:** Reports illegal characters, malformed numbers

#### 2. **Syntax Analysis (Parsing)**
- **What it does:** Takes tokens from lexical analyzer and checks if they follow the grammar rules of the language
- **Output:** Abstract Syntax Tree (AST) or Parse Tree showing hierarchical structure
- **Example:** Verifies that `if` statement has proper condition and body
- **Error detection:** Missing semicolons, unmatched parentheses, wrong statement structure
- **Result:** IF syntax correct, builds parse tree; IF incorrect, reports error

#### 3. **Semantic Analysis**
- **What it does:** Checks meaning and consistency of the program
- **Tasks performed:**
  - **Type checking:** Ensure operations are valid (can't add string + number)
  - **Scope resolution:** Verify variables declared before use
  - **Symbol table management:** Track declarations and their types
  - **Type conversions:** Promote int to float if needed
- **Example:** `int x = 5; x = x + "hello";` → ERROR (can't add int + string)

#### 4. **Intermediate Code Generation**
- **What it does:** Converts high-level code to intermediate representation (IR)
- **Why needed:** 
  - Makes optimization easier
  - Separates front-end (parsing) from back-end (code generation)
  - Same IR can target multiple architectures
- **Forms of IR:** Three-Address Code (TAC), Abstract Syntax Trees, DAGs (Directed Acyclic Graphs)
- **Example:**
  - Source: `a = b + c * d`
  - IR: `t1 = c * d; t2 = b + t1; a = t2;`

#### 5. **Code Optimization**
- **Machine-independent optimizations:** Work on IR, valid for any target
  - Constant folding: `5 * 3` → `15`
  - Dead code elimination: Remove unreachable code
  - Loop invariant code motion: Move expressions outside loops
  - Common subexpression elimination: Reuse computed values
- **Machine-dependent optimizations:** Use target architecture knowledge
  - Register allocation: Assign variables to registers efficiently
  - Instruction scheduling: Reorder to avoid pipeline stalls
  - Strength reduction: Replace expensive ops with cheaper ones
- **Goal:** Faster, smaller, or more efficient code

#### 6. **Code Generation**
- **What it does:** Converts optimized IR to target machine code
- **Tasks:**
  - Select appropriate machine instructions for each IR statement
  - Allocate registers/memory locations for variables
  - Generate addressing modes (immediate, register, indirect, etc.)
  - Ensure calling conventions are followed
- **Output:** Assembly code or binary machine code
- **Example:**
  - IR: `t1 = a + b`
  - Assembly: `LOAD R1, a; LOAD R2, b; ADD R1, R1, R2`

#### 7. **Assembly & Linking**
- **Assembly phase:** Converts assembly mnemonics to machine code (binary)
- **Linking phase:** Combines multiple object files and libraries
- **Tasks:** Resolve external references, relocate addresses, finalize executable

**Summary Table:**

| Phase | Input | Output | Main Task |
|-------|-------|--------|----------|
| Lexical | Source code | Tokens | Group chars into tokens |
| Syntax | Tokens | Parse tree | Check grammar |
| Semantic | Parse tree | Annotated tree | Type check, scope |
| IR Generation | Annotated tree | TAC/IR | Create intermediate |
| Optimization | IR | Optimized IR | Improve efficiency |
| Code Gen | IR | Assembly | Generate target code |
| Assembly | Assembly | Object file | Binary conversion |

## Q2: Differentiate between Compiler and Interpreter with detailed comparison.

### DETAILED ANSWER:

**Detailed Comparison Table:**

| Aspect | Compiler | Interpreter |
|--------|----------|-------------|
| **Definition** | Translates entire program to machine code before execution | Translates and executes line-by-line during runtime |
| **Processing Model** | Two-phase: compilation then execution | Single phase: translation during execution |
| **Translation Time** | Done once at compile-time (takes time upfront) | Done repeatedly at run-time |
| **Execution Speed** | Fast (no translation overhead during run) | Slow (repeated parsing/translation of same code) |
| **Memory Usage** | High (entire compiled program in memory) | Low (only current statement in memory) |
| **Debugging** | Difficult (source not available at runtime) | Easy (can inspect source, step through) |
| **Portability** | Compiled code specific to one target | Source code portable, interpreter needed on each platform |
| **Error Detection** | All errors caught before execution | Errors found only when that line executes |
| **Startup Time** | Slower (must compile first) | Faster (immediate start) |
| **Examples** | C, C++, Java (bytecode), Go | Python, JavaScript, Lisp, Ruby |

**When to Use Compiler:**
1. **Performance-critical applications** – Gaming, real-time systems, simulations
2. **Embedded systems** – Limited memory/CPU, no runtime overhead allowed
3. **Large projects** – Security, maintainability benefits
4. **Scientific computing** – Complex calculations requiring speed

**When to Use Interpreter:**
1. **Scripting & automation** – Shell scripts, automation tasks
2. **Rapid prototyping** – Quick testing without compilation
3. **Learning & education** – Immediate feedback, interactive
4. **Dynamic languages** – Features require runtime flexibility
5. **Cross-platform tools** – Write once, run anywhere with interpreter

**Hybrid Approach (Just-In-Time Compilation):**
- **Java model:** Source → Bytecode (compiler) → JVM interprets → JIT compiles hot code
- **Python model:** Source → Bytecode (compiler) → CPython interpreter
- **Benefits:** Portability of interpreter + speed of compiled code

## Q3: Explain Two-Pass Assembler in detail. What is a symbol table? Why two passes?

### DETAILED ANSWER:

#### What is an Assembler?
An assembler is a program that converts assembly language (mnemonics) into machine code. Unlike compilers (which translate high-level languages), assemblers perform a simpler one-to-one mapping of assembly instructions to machine instructions.

#### Why Two Passes?
**The Problem of Forward References:**
```
JNE  LOOP        ← Uses label LOOP before it's defined (forward reference)
LD   A, 5
LOOP: ADD A, B   ← Label LOOP defined here
```

To generate correct machine code for the JNE instruction, we need to know the address of LOOP. But LOOP hasn't been encountered yet! This is why we need two passes:
- **Pass 1:** Collect all label addresses
- **Pass 2:** Generate code using known addresses

#### Pass 1: Symbol Definition & Table Construction

**What happens in Pass 1:**

1. Initialize location counter (LC) = 0
2. Scan source assembly code line by line
3. For each **label** found:
   - Add to symbol table: `label_name → current_LC_value`
4. For each **instruction**:
   - Increment LC by instruction size (usually 2-4 bytes)
5. Ignore non-label, non-instruction lines (comments, directives except size allocation)
6. Output: Complete symbol table

**Example Pass 1:**
```
Line   Code              Label?   LC Before   LC After   Symbol Table
1      START:LD R1,10    YES      0x0000      0x0002     START → 0x0000
2      ADD R1,R2         NO       0x0002      0x0004     
3      LOOP: MUL R1,2    YES      0x0004      0x0006     LOOP → 0x0004
4      CMP R1,100        NO       0x0006      0x0008     
5      JNE LOOP          NO       0x0008      0x000A     
6      ST R1,RESULT      NO       0x000A      0x000C     
7      HLT               NO       0x000C      0x000E     
8      RESULT: (space)   YES      0x000E      0x0012     RESULT → 0x000E
```

**Final Symbol Table from Pass 1:**
```
| Symbol | Address | Type   | Scope |
|--------|---------|--------|-------|
| START  | 0x0000  | Label  | Local |
| LOOP   | 0x0004  | Label  | Local |
| RESULT | 0x000E  | Data   | Local |
```

#### Pass 2: Code Generation Using Symbol Table

**What happens in Pass 2:**

1. Initialize LC = 0 again
2. Scan source assembly code again
3. For each instruction:
   - If instruction contains label reference → lookup in symbol table → substitute address
   - Generate machine code for instruction
   - Output to object file
   - Increment LC
4. Output: Machine code (object file)

**Example Pass 2:**
```
Address  Instruction      Operand Lookup         Machine Code
0x0000   LD R1, 10        -                       10 0A
0x0002   ADD R1, R2       -                       20 02
0x0004   MUL R1, 2        -                       30 02
0x0006   CMP R1, 100      -                       40 64
0x0008   JNE LOOP         lookup LOOP → 0x0004   50 04 ✓ (resolved!)
0x000A   ST R1, RESULT    lookup RESULT → 0x000E 60 0E ✓ (resolved!)
0x000C   HLT              -                       FF FF
```

#### Symbol Table: Complete Detail

**Purpose:** Map symbolic names (labels, variables) to their addresses

**Structure (implementation example):**
```
typedef struct {
    char name[32];        // Label/variable name
    int address;          // Memory address
    char type[16];        // "Label", "Variable", "Data"
    char scope[16];       // "Global", "Local", "External"
    int size;             // For data (bytes allocated)
} SymbolEntry;
```

**Operations on Symbol Table:**
1. **INSERT:** Add new entry during Pass 1
2. **LOOKUP:** Find address during Pass 2
3. **PRINT:** Display all symbols at end

#### Summary Comparison

| Aspect | Pass 1 | Pass 2 |
|--------|--------|--------|
| **Goal** | Build symbol table | Generate code |
| **Input** | Assembly source | Assembly source + symbol table |
| **Output** | Symbol table | Machine code |
| **Handles** | Label definitions | Label references |
| **Forward refs?** | No (just collect labels) | Yes (uses table) |

### PASS 1: SYMBOL TABLE CONSTRUCTION

**Step-by-step trace:**

```
Initialize: LC = 0x0000, Symbol Table = empty

Line 1: START: LD R1, 10
  → Label "START" found at LC = 0x0000
  → Add to table: START → 0x0000
  → Instruction size: 2 bytes
  → LC = 0x0000 + 2 = 0x0002

Line 2: ADD R1, R2
  → No label
  → Instruction size: 2 bytes
  → LC = 0x0002 + 2 = 0x0004

Line 3: LOOP: MUL R1, 2
  → Label "LOOP" found at LC = 0x0004
  → Add to table: LOOP → 0x0004
  → Instruction size: 2 bytes
  → LC = 0x0004 + 2 = 0x0006

Line 4: CMP R1, 100
  → No label
  → Instruction size: 2 bytes
  → LC = 0x0006 + 2 = 0x0008

Line 5: JNE LOOP
  → No label (this is an instruction with forward reference)
  → Instruction size: 2 bytes
  → LC = 0x0008 + 2 = 0x000A

Line 6: ST R1, RESULT
  → No label
  → Instruction size: 2 bytes
  → LC = 0x000A + 2 = 0x000C

Line 7: HLT
  → No label
  → Instruction size: 2 bytes
  → LC = 0x000C + 2 = 0x000E

Line 8: RESULT: .SPACE 4
  → Label "RESULT" found at LC = 0x000E
  → Add to table: RESULT → 0x000E
  → Directive: allocate 4 bytes
  → LC = 0x000E + 4 = 0x0012
```

### FINAL SYMBOL TABLE (Pass 1 Output):

```
┌─────────────────────────────────────────┐
│        SYMBOL TABLE                     │
├─────────────┬──────────┬──────┬─────────┤
│ Symbol Name │ Address  │ Type │ Scope   │
├─────────────┼──────────┼──────┼─────────┤
│ START       │ 0x0000   │ Label│ Local   │
│ LOOP        │ 0x0004   │ Label│ Local   │
│ RESULT      │ 0x000E   │ Data │ Local   │
└─────────────┴──────────┴──────┴─────────┘
```

---

### PASS 2: CODE GENERATION WITH RESOLVED REFERENCES

**Step-by-step code generation:**

```
Initialize: LC = 0x0000 again
Use Symbol Table from Pass 1 to resolve references

┌──────┬──────────────────┬──────────────────┬──────────────┐
│ Addr │ Instruction      │ Resolution       │ Machine Code │
├──────┼──────────────────┼──────────────────┼──────────────┤
│0x0000│ LD R1, 10        │ No refs          │ 10 0A        │
│      │ (START)          │ Opcode:10,Op:0A  │              │
├──────┼──────────────────┼──────────────────┼──────────────┤
│0x0002│ ADD R1, R2       │ No refs          │ 20 02        │
│      │                  │ Opcode:20,Reg:02 │              │
├──────┼──────────────────┼──────────────────┼──────────────┤
│0x0004│ MUL R1, 2        │ No refs          │ 30 02        │
│      │ (LOOP)           │ Opcode:30,Op:02  │              │
├──────┼──────────────────┼──────────────────┼──────────────┤
│0x0006│ CMP R1, 100      │ No refs          │ 40 64        │
│      │                  │ Opcode:40,Op:64  │              │
├──────┼──────────────────┼──────────────────┼──────────────┤
│0x0008│ JNE LOOP         │ LOOP lookup→0x04 │ 50 04        │
│      │ (forward ref!)   │ Opcode:50,Addr:04│ ✓ RESOLVED!  │
├──────┼──────────────────┼──────────────────┼──────────────┤
│0x000A│ ST R1, RESULT    │ RESULT lookup→0E │ 60 0E        │
│      │                  │ Opcode:60,Addr:0E│ ✓ RESOLVED!  │
├──────┼──────────────────┼──────────────────┼──────────────┤
│0x000C│ HLT              │ No refs          │ FF FF        │
│      │                  │ Opcode:FF        │              │
└──────┴──────────────────┴──────────────────┴──────────────┘
```

### FINAL OBJECT CODE (Pass 2 Output):

```
Memory Address | Machine Code (Hex) | Disassembly
───────────────┼───────────────────┼─────────────────────────
0x0000         | 10 0A             | LD R1, 10
0x0002         | 20 02             | ADD R1, R2
0x0004         | 30 02             | MUL R1, 2
0x0006         | 40 64             | CMP R1, 100
0x0008         | 50 04             | JNE 0x0004 (LOOP resolved!)
0x000A         | 60 0E             | ST R1, 0x000E (RESULT resolved!)
0x000C         | FF FF             | HLT
0x000E-0x0011  | (data space)      | (4 bytes for RESULT)
```

### KEY OBSERVATIONS:

1. **Forward Reference Resolution:**
   - Line 5 (0x0008): `JNE LOOP` → LOOP at 0x0004 → Machine code: 50 04 ✓
   - Line 6 (0x000A): `ST R1, RESULT` → RESULT at 0x000E → Machine code: 60 0E ✓
   - Without Pass 1, we wouldn't know 0x0004 and 0x000E!

2. **Two-Pass Efficiency:**
   - Pass 1 time: O(n) where n = number of lines
   - Pass 2 time: O(n)
   - Total: O(2n) = O(n) — linear time, very efficient

3. **Why Not One Pass?**
   - If we tried to generate code on first pass, we'd need to backpatch (leave holes) when we encounter forward references
   - More complex, less elegant than two-pass approach

## Q5: Backpatching in One-Pass Assembler (Alternative approach)

### DETAILED EXPLANATION:

Some assemblers use backpatching to avoid two passes. Here's how:

**Forward Reference Table (FRT):**
```
When JNE LOOP encountered at 0x0008:
  → LOOP not yet defined
  → Add to FRT: (LOOP, 0x0008, pending)
  → Output placeholder machine code: 50 00 (unknown address)
```

**When Label Found:**
```
When LOOP: found at 0x0004:
  → Search FRT for "LOOP"
  → Found: reference at 0x0008
  → Go back to 0x0008, update machine code to 50 04
  → Mark as resolved
```

**Pros:** Only one pass  
**Cons:** More complex (need file seeking/buffering)

**Most modern assemblers use two-pass approach (simpler, faster)**

## Q1: Define macro and macro processor. Differentiate macro from subroutine.

### DETAILED ANSWER:

#### Definition: Macro

A **macro** is a named sequence of instructions that can be invoked multiple times with different parameters. The macro **body is expanded inline** at each call site, replacing formal parameters with actual arguments.

**Simple Example:**
```
MACRO ADD_TO_ACC(value)
    ADD ACC, value
ENDM

Usage:
  ADD_TO_ACC(5)   → Expands to: ADD ACC, 5
  ADD_TO_ACC(10)  → Expands to: ADD ACC, 10
  ADD_TO_ACC(R1)  → Expands to: ADD ACC, R1
```

#### Definition: Macro Processor

A **macro processor** is a program that:
1. Recognizes macro definitions in source code
2. Stores macro bodies for later use
3. When macro is called, retrieves body and expands it
4. Substitutes actual parameters for formal parameters
5. Outputs expanded text

**Note:** Macro processors can be standalone (independent) or part of assembler/compiler.

#### Macro vs Subroutine: Detailed Comparison

| Characteristic | Macro | Subroutine |
|---|---|---|
| **Mechanism** | Text substitution (preprocessing) | Runtime call/return |
| **When processed** | Assembly/compile time | Runtime |
| **Expansion** | Code duplicated at each call site | Shared code, one location |
| **Overhead** | NONE (expansion happens before execution) | CALL/RET overhead |
| **Code size** | INCREASES with each macro call | CONSTANT (one definition) |
| **Execution speed** | FASTER (no call overhead) | SLOWER (branch + stack ops) |
| **Debugging** | Expanded code, harder to trace | Can debug call/return flow |
| **Parameter passing** | Text substitution, very flexible | Formal: registers, stack, etc. |
| **Best for** | Small, frequently-used, time-critical | Large or occasional code |

#### Example Comparison

**Using Macro:**
```
MACRO SWAP(A, B, T)
    MOV T, A
    MOV A, B
    MOV B, T
ENDM

Code:
    SWAP(R1, R2, R3)    ; Call 1
    ...
    SWAP(R4, R5, R6)    ; Call 2

After Macro Expansion (code duplication!):
    MOV R3, R1          ; Expansion 1
    MOV R1, R2
    MOV R2, R3
    ...
    MOV R6, R4          ; Expansion 2
    MOV R4, R5
    MOV R5, R6
```

**Using Subroutine:**
```
SWAP:
    MOV T, A            ; Shared code
    MOV A, B
    MOV B, T
    RET

Code:
    CALL SWAP           ; Call 1 (branch to SWAP)
    ...
    CALL SWAP           ; Call 2 (branch to same SWAP)

Result: One copy of SWAP code, two branches to it
```

#### When to Use Each

**Use MACRO when:**
- Small code (2-10 instructions)
- Called frequently
- Time-critical sections
- Need flexibility with parameters

**Use SUBROUTINE when:**
- Large code (20+ instructions)
- Called occasionally
- Code reuse important
- Memory efficiency critical

**Required:** Show the complete macro expansion for all calls, with step-by-step parameter substitution.

### SOLUTION:

---

### MACRO EXPANSION PROCESS

#### Step 1: Parse Macro Definition
```
Macro name: SWAP
Formal parameters: A, B, TEMP (3 parameters)
Macro body:
  Line 1: MOV TEMP, A
  Line 2: MOV A, B
  Line 3: MOV B, TEMP
Status: SWAP definition stored, ready for use
```

#### Step 2: First Macro Call - SWAP(R1, R2, R3)

```
Call encountered: SWAP(R1, R2, R3)

Parameter Mapping:
  Formal A    ← Actual R1
  Formal B    ← Actual R2
  Formal TEMP ← Actual R3

Macro Body Substitution:
  Original:     MOV TEMP, A
  Substitute:   MOV R3, R1     (TEMP→R3, A→R1)
  
  Original:     MOV A, B
  Substitute:   MOV R1, R2     (A→R1, B→R2)
  
  Original:     MOV B, TEMP
  Substitute:   MOV R2, R3     (B→R2, TEMP→R3)

Expanded Code for First Call:
  MOV R3, R1
  MOV R1, R2
  MOV R2, R3
```

#### Step 3: Second Macro Call - SWAP(X, Y, TEMP_VAR)

```
Call encountered: SWAP(X, Y, TEMP_VAR)

Parameter Mapping:
  Formal A    ← Actual X
  Formal B    ← Actual Y
  Formal TEMP ← Actual TEMP_VAR

Macro Body Substitution:
  Original:     MOV TEMP, A
  Substitute:   MOV TEMP_VAR, X
  
  Original:     MOV A, B
  Substitute:   MOV X, Y
  
  Original:     MOV B, TEMP
  Substitute:   MOV Y, TEMP_VAR

Expanded Code for Second Call:
  MOV TEMP_VAR, X
  MOV X, Y
  MOV Y, TEMP_VAR
```

#### Final Expanded Source Code:

```
Original source:
    SWAP(R1, R2, R3)
    ADD  R1, 1
    SWAP(X, Y, TEMP_VAR)
    JNE  LOOP

After Macro Expansion:
    MOV R3, R1          ┐
    MOV R1, R2          ├─ SWAP(R1, R2, R3) expanded
    MOV R2, R3          ┘
    ADD R1, 1           ← Not a macro, unchanged
    MOV TEMP_VAR, X     ┐
    MOV X, Y            ├─ SWAP(X, Y, TEMP_VAR) expanded
    MOV Y, TEMP_VAR     ┘
    JNE LOOP            ← Not a macro, unchanged
```

### KEY POINTS:

1. **Code Duplication:** SWAP body appears TWICE (once for each call)
2. **Parameter Substitution:** Each parameter consistently replaced throughout body
3. **Different Arguments:** Same macro with different arguments → different expansions
4. **Assembly Time:** Expansion happens during macro preprocessing, before actual assembly
5. **No Runtime Overhead:** Expanded code executes directly, no branching

## Q3: Macro Processor Data Structures: MNT, MDT, ALA (In Depth)

### DETAILED EXPLANATION:

Macro processors maintain three critical data structures to manage macro definitions and expansions:

#### 1. MACRO NAME TABLE (MNT)

**Purpose:** Quick lookup of macro names and their metadata

**Structure:**
```
Each entry contains:
  • Macro name (key for lookup)
  • Index into MDT (where macro body starts)
  • Number of formal parameters
  • List of formal parameter names
```

**Example MNT:**
```
┌─────────────────────────────────────────────────────────────┐
│                    MACRO NAME TABLE (MNT)                   │
├─────────┬──────────┬─────────┬─────────────────────────────┤
│ Name    │ MDT_Idx  │ #Params │ Formal Parameters           │
├─────────┼──────────┼─────────┼─────────────────────────────┤
│ SWAP    │ 0        │ 3       │ A, B, TEMP                  │
│ PRINT   │ 5        │ 2       │ FORMAT, VALUE               │
│ MAX     │ 12       │ 2       │ X, Y                        │
│ MIN     │ 18       │ 2       │ X, Y                        │
└─────────┴──────────┴─────────┴─────────────────────────────┘
```

**Example lookup:**
- User calls `SWAP(...)`
- Search MNT for "SWAP" → Found at row 1
- Get MDT_Idx = 0 (body starts at MDT line 0)
- Get #Params = 3
- Get formal param names = [A, B, TEMP]

#### 2. MACRO DEFINITION TABLE (MDT)

**Purpose:** Store complete body of each macro, line by line

**Structure:**
```
Each entry:
  • Line number (0, 1, 2, ...)
  • Instruction/text of line
  • Contains formal parameter names (will be substituted later)
```

**Example MDT:**
```
┌────────┬────────────────────────────────────────────────────────┐
│ MDT_Idx│ Macro Body Line                                        │
├────────┼────────────────────────────────────────────────────────┤
│ 0      │ MOV TEMP, A                  ← SWAP macro body starts  │
│ 1      │ MOV A, B                                               │
│ 2      │ MOV B, TEMP                                            │
│ 3      │ ENDM                         ← SWAP macro body ends    │
│ 4      │ (empty)                                                │
│ 5      │ PUSH FORMAT                  ← PRINT macro body starts │
│ 6      │ PUSH VALUE                                             │
│ 7      │ CALL SYS_PRINT                                         │
│ 8      │ ADD SP, 8                                              │
│ 9      │ ENDM                         ← PRINT macro body ends   │
│ 10     │ (empty)                                                │
│ 11     │ (other macros...)                                      │
└────────┴────────────────────────────────────────────────────────┘
```

**Why MDT is organized this way:**
- Each macro body stored consecutively
- MNT contains starting index (MDT_Idx)
- ENDM marker signals end of macro
- Avoids duplication of text (one definition, referenced from MNT)

#### 3. ARGUMENT LIST ARRAY (ALA)

**Purpose:** Temporarily store actual arguments when macro is called, for parameter substitution

**Structure:**
```
Each entry:
  • Parameter position (1, 2, 3, ...)
  • Actual argument (text passed in call)
  
ALA is ephemeral: created for each macro call, destroyed after expansion
```

**Example ALA When SWAP(R1, R2, R3) Called:**
```
┌─────────────┬────────────┐
│ Position    │ Argument   │
├─────────────┼────────────┤
│ ALA[1]      │ R1         │ (formal A → actual R1)
│ ALA[2]      │ R2         │ (formal B → actual R2)
│ ALA[3]      │ R3         │ (formal TEMP → actual R3)
└─────────────┴────────────┘
```

**Example ALA When SWAP(X, Y, TEMP_VAR) Called:**
```
┌─────────────┬────────────┐
│ Position    │ Argument   │
├─────────────┼────────────┤
│ ALA[1]      │ X          │
│ ALA[2]      │ Y          │
│ ALA[3]      │ TEMP_VAR   │
└─────────────┴────────────┘
```

### MACRO EXPANSION ALGORITHM (Using These Structures)

```
ALGORITHM MacroExpand(macro_name, actual_args)

1. LOOKUP macro_name in MNT
   IF not found:
     Error: Undefined macro
     RETURN
   ELSE:
     mdtIndex ← MNT entry's MDT_Idx
     paramCount ← MNT entry's #Params
     formalParams ← MNT entry's Formal Parameter List

2. BUILD ALA from actual arguments
   FOR i = 1 TO paramCount:
     ALA[i] ← actual_args[i]

3. EXPAND macro body
   lineNum ← mdtIndex
   WHILE MDT[lineNum] != "ENDM":
     line ← MDT[lineNum]
     
     // Substitute all formal parameters with actual arguments
     FOR i = 1 TO paramCount:
       line ← Replace(line, formalParams[i], ALA[i])
     
     OUTPUT line
     lineNum ← lineNum + 1

4. CLEANUP (ALA destroyed after expansion)
   Clear ALA
   
END
```

### Example: Complete Walkthrough

```
User writes: SWAP(R1, R2, R3)

Step 1: Lookup "SWAP" in MNT
  → Found: MDT_Idx=0, #Params=3, Params=[A,B,TEMP]

Step 2: Build ALA
  ALA[1] = R1
  ALA[2] = R2
  ALA[3] = R3

Step 3: Expand body
  MDT[0]: "MOV TEMP, A"
    Replace A with ALA[1]=R1: "MOV TEMP, R1"
    Replace TEMP with ALA[3]=R3: "MOV R3, R1"
    Output: MOV R3, R1
  
  MDT[1]: "MOV A, B"
    Replace A with R1, B with R2: "MOV R1, R2"
    Output: MOV R1, R2
  
  MDT[2]: "MOV B, TEMP"
    Replace B with R2, TEMP with R3: "MOV R2, R3"
    Output: MOV R2, R3
  
  MDT[3]: "ENDM" → stop

Result:
  MOV R3, R1
  MOV R1, R2
  MOV R2, R3
```

## Q1: Lexical Analysis - Tokens, Lexemes, Patterns (Detailed)

### DETAILED EXPLANATION:

**Lexical Analysis** is the first phase of compilation. It converts raw source code into a sequence of tokens (symbolic units). This simplifies parsing and separates character-level syntax from grammar-level syntax.

#### Three Core Concepts:

### 1. LEXEME

**Definition:** The actual sequence of characters from the source code that form a token.

**Examples:**
- `int` (keyword)
- `count` (identifier)
- `42` (numeric literal)
- `+` (operator)
- `"hello"` (string literal)

**Key point:** Lexeme is the ACTUAL TEXT from source, not an abstraction.

### 2. TOKEN

**Definition:** A symbolic name representing a class of lexemes.

**Token Format:** `<token-name, attribute-value>`

**Examples:**
- Lexeme `int` → Token `<KEYWORD, int>`
- Lexeme `count` → Token `<IDENTIFIER, count>`
- Lexeme `42` → Token `<NUMBER, 42>`
- Lexeme `+` → Token `<OPERATOR, +>`

**Why tokens?** Parser works with abstract token names, not text. Simplifies grammar rules.

### 3. PATTERN

**Definition:** A rule (usually regular expression) that specifies which lexemes belong to a particular token class.

**Examples:**

| Token | Pattern | Example Lexemes |
|-------|---------|----------------|
| IDENTIFIER | `[a-zA-Z][a-zA-Z0-9_]*` | count, x, _temp, name42 |
| NUMBER | `[0-9]+` | 0, 42, 999 |
| FLOAT | `[0-9]+\.[0-9]+` | 3.14, 0.5 |
| STRING | `"[^"]*"` | "hello", "test" |
| KEYWORD | `int\|if\|while\|for` | int, if, while, for |
| OPERATOR | `+\|-\|*\|/` | +, -, *, / |

#### Complete Example:

**Source code:**
```c
int count = 5 + 3;
```

**Lexical analysis:**

```
Lexeme Stream:
  "int"    "count"   "="    "5"    "+"    "3"    ";"

Token Stream (output of lexer):
  <KEYWORD, int>
  <IDENTIFIER, count>
  <ASSIGN, =>
  <NUMBER, 5>
  <PLUS, +>
  <NUMBER, 3>
  <SEMICOLON, ;>
```

#### Lexer Implementation (High Level)

```
ALGORITHM Lexer(source_code)

1. Initialize:
   position ← 0
   tokens ← empty list
   keywords ← {int, if, while, for, ...}

2. WHILE position < length(source_code):

   a) Skip whitespace:
      IF source[position] is space/tab/newline:
        position ← position + 1
        continue

   b) Try to match each pattern:
      IF matches NUMBER pattern:
        lexeme ← extract number
        tokens.append(<NUMBER, lexeme>)
      
      ELSE IF matches IDENTIFIER/KEYWORD pattern:
        lexeme ← extract identifier
        IF lexeme in keywords:
          tokens.append(<KEYWORD, lexeme>)
        ELSE:
          tokens.append(<IDENTIFIER, lexeme>)
      
      ELSE IF matches OPERATOR pattern:
        lexeme ← extract operator
        tokens.append(<OPERATOR, lexeme>)
      
      ELSE IF matches STRING pattern:
        lexeme ← extract string
        tokens.append(<STRING, lexeme>)
      
      ELSE:
        Error: Illegal character

3. RETURN tokens

END
```

#### Why Lexical Analysis Matters

1. **Separation of Concerns:** Character-level syntax separate from grammar
2. **Performance:** Regex matching faster than full parser
3. **Error Handling:** Catch lexical errors early (illegal chars, malformed numbers)
4. **Symbiotic Support:** Tracks line/column numbers for error messages

## Q2: NUMERICAL - Regular Expression to DFA (Complete Derivation)

### PROBLEM:

Convert the regular expression `(a|b)*abb` to a Deterministic Finite Automaton (DFA). This regex accepts strings with any combination of 'a' and 'b', ending exactly with 'abb'.

**Examples of accepted strings:**
- `abb` ✓
- `aabb` ✓ 
- `babb` ✓
- `aaabb` ✓
- `baababb` ✓

**Examples of rejected strings:**
- `ab` ✗ (missing final 'b')
- `abba` ✗ (extra character)
- `aab` ✗ (doesn't end with 'abb')

### SOLUTION:

---

### STEP 1: Understand the Regular Expression

```
Regex: (a|b)*abb

Breaking it down:
  (a|b)*  = Zero or more occurrences of 'a' OR 'b' (loop part)
  abb     = Exactly the string 'abb' (final part)

Total pattern: [Loop part] followed by [Final 'abb']
```

### STEP 2: Build NFA (Non-deterministic Finite Automaton) Using Thompson's Algorithm

**Thompson's Algorithm constructs NFA from regex by combining smaller NFAs.**

```
For (a|b): Two paths
   q0 -a→ q1
   q0 -b→ q2

For (a|b)*: Add epsilon loops
   Entry -ε→ (a|b) NFA -ε→ Exit
   Entry -ε→ Exit (empty, for zero repetitions)

Complete (a|b)*abb NFA:
   States: q0 (start) to q10 (accept)
   
   q0 -ε→ q1                    (enter loop)
   q0 -ε→ q7                    (skip loop)
   q1 -ε→ q2 (for 'a' branch)
   q1 -ε→ q4 (for 'b' branch)
   q2 -a→ q3
   q4 -b→ q5
   q3 -ε→ q6
   q5 -ε→ q6
   q6 -ε→ q1 (loop back for *)
   q6 -ε→ q7 (exit loop)
   q7 -a→ q8                    (first 'a' of 'abb')
   q8 -b→ q9                    (first 'b' of 'abb')
   q9 -b→ q10 (accept)          (final 'b')
```

### STEP 3: Compute Epsilon Closures for Each NFA State

**ε-closure(q) = set of states reachable from q using only ε-transitions**

```
ε-closure(q0) = {q0, q1, q2, q4, q7}        (can reach all loop entry/start states)
ε-closure(q1) = {q1, q2, q4}
ε-closure(q2) = {q2}
ε-closure(q3) = {q3, q6, q1, q2, q4, q7}    (epsilon back to loop)
ε-closure(q4) = {q4}
ε-closure(q5) = {q5, q6, q1, q2, q4, q7}
ε-closure(q6) = {q6, q1, q2, q4, q7}
ε-closure(q7) = {q7}
ε-closure(q8) = {q8}
ε-closure(q9) = {q9}
ε-closure(q10) = {q10}
```

### STEP 4: NFA to DFA Conversion (Subset Construction)

**Algorithm:** Each DFA state = subset of NFA states

```
DFA state q0 = ε-closure(NFA q0) = {q0,q1,q2,q4,q7}

From q0, on input 'a':
  From q2 (in closure of q0): q2 -a→ q3
  ε-closure(q3) = {q3,q6,q1,q2,q4,q7}
  → DFA state q1 = {q3,q6,q1,q2,q4,q7} (can loop or enter 'abb' part)
  
  Also from q7 (in closure of q0): q7 -a→ q8
  ε-closure(q8) = {q8}
  → Merged: DFA state q1' = {q3,q6,q1,q2,q4,q7,q8}

From q0, on input 'b':
  From q4 (in closure of q0): q4 -b→ q5
  ε-closure(q5) = {q5,q6,q1,q2,q4,q7}
  → DFA state q2 = {q5,q6,q1,q2,q4,q7}
  
Continue building transitions...
```

### STEP 5: Simplified DFA Transition Table

```
DFA State    Input: a         Input: b
────────────────────────────────────────
q0 (start)   q1               q2
q1           q1               q3
q2           q1               q2
q3 (accept)  -                -
```

**Explanation:**
- `q0`: Initial state (in loop part)
- `q1`: After seeing 'a' in current position
- `q2`: After seeing 'b' in current position (not final)
- `q3`: After seeing 'abb' sequence (accept state)

### STEP 6: DFA State Diagram

```
           a,b
          ↙   ↖
    ┌─→ q0 ──a──→ q1
    │    │         │ b
    │    │ b       ↓
    │    ↓         q3 (accept)
    └─── q2 ←──a── (no outgoing from q3)
         │  ↖
         │───┘ (stay in q2 on 'b')
```

### STEP 7: Test Strings

```
Test: "abb"
  q0 -a→ q1 -b→ q3 -b→? NO!
  Error in my transition table. Let me recalculate...
  
Actually, after 'a' from q0, we go to q1.
From q1, on first 'b': should go to state tracking "ab"
Let me rename:
  q0: Initial (seen 0 chars of 'abb')
  q1: Seen 'a' OR in loop
  q2: Seen 'ab'
  q3: Seen 'abb' (ACCEPT)

Corrected transitions:
  From q0 on 'a': go to q1 (might start 'abb')
  From q0 on 'b': stay in q0 (restart loop)
  From q1 on 'a': stay in q1 (restart 'a' of 'abb')
  From q1 on 'b': go to q2 (have 'ab' so far)
  From q2 on 'a': go to q1 (restart)
  From q2 on 'b': go to q3 (ACCEPT!)
  From q3: no transitions (accept state)

Test: "abb"
  q0 -a→ q1 -b→ q2 -b→ q3 ✓ ACCEPT
  
Test: "aabb"
  q0 -a→ q1 -a→ q1 -b→ q2 -b→ q3 ✓ ACCEPT
  
Test: "ab"
  q0 -a→ q1 -b→ q2 ✗ REJECT (stuck at q2, not q3)
  
Test: "babb"
  q0 -b→ q0 -a→ q1 -b→ q2 -b→ q3 ✓ ACCEPT
```
